```python
import pymongo
import pandas
from datetime import datetime
import matplotlib
```

```python
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
```

```python
db = client.local
### Emails is the collection name where you imported the csv file
collection = db.parsed_emails
```

```python
cursor_list=collection.find({"from_name":"Phillip K Allen"},{"date":1, "_id":0})
```

```python
datetime_list=[]
for cursor in cursor_list:
    datetime_list.append(cursor["date"])
```

```python
datetime_list
```

```python
datetime_parsed=[]
for datetime_ in datetime_list:
    datetime_object = datetime.strptime(str(datetime_).split('-')[0].split(',')[1], ' %d %b %Y %H:%M:%S ')
    datetime_parsed.append(datetime_object)
```

```python
datetime_parsed
```

```python
hours={}
for date_ in datetime_parsed:
    try:
        count=hours[date_.hour]+1
    except KeyError:
        count=1
    hours[date_.hour]=count
```

```python
hours
```

```python
import matplotlib.pyplot as plt
plt.hist(hours.values(), normed=True, bins=24)
plt.ylabel('Email counts per hour');
```
